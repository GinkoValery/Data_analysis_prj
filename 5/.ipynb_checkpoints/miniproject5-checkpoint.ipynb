# Дополнительный проект
В папке subsid (5_subsid) находятся файлы (tm_sales_1, tm_sales_2, ...) с продажами продуктов через телемаркетинг. Каждый файл содержит, как минимум, 4 колонки (поля): FILIAL_ID, SUBS_ID, PROD_ID, ACT_DTTM.
Суть задачи в том, чтобы проверить подключения продуктов определенным пользователям, соединив файлы о продажах с логами по подключениям в системе.
Особенности данных:
1.	сотрудники телемаркетинга не всегда указывают полный id, если 'id' нет в начале SUBS_ID, то нужно его добавить
2.	поля в файлах могут быть расположены абсолютно случайным образом, но названия полей статичны
3.	продажа не засчитывается, если отключение (END_DTTM) произошло меньше чем через 5 минут после подключения (START_DTTM)
4.	если в файле с продажами встречается строка без указанного SUBS_ID, она пропускается


```python
from pathlib import Path
```

```python
def get_data_from_folder(path):
    df = pd.read_csv(path, sep = ";")
    cols = list(filter(lambda x: "DTTM" in x, df.columns))
    df[cols] = df[cols].apply(lambda x: pd.to_datetime(x, format='%d-%m-%Y %H:%M'))
    return df
```

```python
pth = Path(Path.cwd(), "5_subsid")
pth
```

```python
# Запишем в дф данные по продажам из всех файлов
dfs = [get_data_from_folder(f) for f in pth.glob("*_sales_*")]
sales = pd.concat(dfs, sort = False)
sales
```

```python
sales.info()
```

```python
# если в файле с продажами встречается строка без указанного SUBS_ID, она пропускается
sales.dropna(subset = ["SUBS_ID"], inplace = True)
sales
```

```python
sales.info()
```

```python
# сотрудники телемаркетинга не всегда указывают полный id, если 'id' нет в начале SUBS_ID, то нужно его добавить
sales.SUBS_ID = sales.SUBS_ID.apply(lambda x: x if str(x).startswith('id') else f"id{x}")
sales
```

```python
# Запишем в дф данные по логам из всех файлов
dfs_logs = [get_data_from_folder(f) for f in pth.glob("*_logs*")]
logs = pd.concat(dfs_logs, sort = False)
logs
```

```python
logs.info()
```

```python
# продажа не засчитывается, если отключение (END_DTTM) произошло меньше чем через 5 минут после подключения (START_DTTM)
logs["time_diff"] = (logs["END_DTTM"] - logs["START_DTTM"]).dt.total_seconds()
logs
```

```python
logs = logs[logs.time_diff > 300]
logs
```

```python
# Суть задачи в том, чтобы проверить подключения продуктов определенным пользователям, соединив файлы о продажах с логами по подключениям в системе.
fact_table = sales.merge(logs, how = "inner", on = ['SUBS_ID', 'PROD_ID'])
fact_table
```

```python
# Для проверки результатов введите SUBS_ID из полученного датасета в порядке возрастания, через запятую с пробелом. 
", ".join(list(fact_table["SUBS_ID"].sort_values()))
```
